## 가장 많이 사용된 단어 알아보기

### 1. 연설문 불러오기

```python
moon = open('speech_moon.txt', encoding = 'UTF-8').read()
moon
```

```python
### 2. 불필요한 문자 제거하기
import re
moon = re.sub('[^가-힇]', ' ', moon)
moon
```

### 3. 명사 추출하기

```python
# hannanum 만들기
import konlpy
```

```python
hannanum = konlpy.tag.Hannanum()
```

```python
# noun
hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi' : '300',
                    'figure.figsize' : [8, 6],
                    'font.size' : 13,
                    'font.family' : 'sans-serif'})
```

```python
# 연설문에서 명사 추출하기
nouns = hannanum.nouns(moon)
```

```python
nouns
```

```python
# data frame
import pandas as pd
df_word = pd.DataFrame({'word' : nouns})
df_word
```

### 4. 단어 빈도표 만들기

```python
# 글자 수 추가
df_word['count'] = df_word['word'].str.len()
df_word
```

```python
## 두 글자 이상 단어만 남기기
df_word = df_word.query('count >= 2')
df_word.sort_values('count')
```

```python
# 단어 빈도 구하기
df_word = df_word.groupby('word', as_index = False)\
    .agg(n = ('word', 'count')) \
        .sort_values('n', ascending = False)

df_word
```

```python
### 5. 단어 막대 그래프 만들기 - 빈도
top20 = df_word.head(20)
top20
```

```python
import seaborn as sns

import matplotlib.pyplot as plt
plt.rcParams.update({'font.family': 'AppleGothic',
                    'figure.dpi' : '1000',
                    'figure.figsize' : [6.5, 6]})

sns.barplot(data = top20, y = 'word', x = 'n')
```

### 1. wordcloud 패키기 설치

```python
pip install wordcloud
```

### 2. 한글 폰트 설정하기

```python
font = 'AppleGothic.ttf'
```

### 3. 단어와 빈도를 담은 딕셔너리 만들기

```python
# data frame -> Dic
dic_word = df_word.set_index('word').to_dict()['n']
dic_word
```

# 10-2 기사 댓글 텍스트 마이닝

## 가장 많이 사용된 단어 알아보기

### 1. 기사 댓글 불러오기

```python
# data 불러오기
df = pd.read_csv('news_comment_BTS.csv', encoding = 'UTF-8')

# 데이터 파악하기
df.info()
```

### 2. 불필요한 문자 제거하기

```python
# 불필요한 문자 제거하기
df['reply'] = df['reply'].str.replace('[^가-힇]', ' ', regex=True)
df['reply'].head()
```

### 3. 명사 추출하기

```python
# kkma
import konlpy
kkma = konlpy.tag.Kkma()
```

```python
# noun
nouns = df['reply'].apply(kkma.nouns)
```

```python
nouns
```

### 4. 단어 빈도표 만들기

```python
# 한 행에 한 단어 들어가도록 구성
nouns = nouns.explode()
nouns
```

```python
# make data frame
df_word = pd.DataFrame({'word' : nouns})

# add word count
df_word['count'] = df_word['word'].str.len()

# word >=2
df_word = df_word.query('count >= 2')
df_word
```

```python
# make
df_word = df_word.groupby('word', as_index = False) \
                            .agg(n = ('word', 'count'))\
                            .sort_values('n', ascending=False)

df_word
```

### 5. 단어 빈도 막대 그래프 만들기

```python
# 단어 빈도 상위 20개 추출
top20 = df_word.head(20)
top20
```

```python
# 가로 크기 설정
plt.rcParams.update({'font.family': 'AppleGothic',
                    'figure.dpi' : '1000',
                    'figure.figsize' : [6.5, 6]})


# 막대 그래프 만들기
sns.barplot(data = top20, y = 'word', x = 'n')
```
